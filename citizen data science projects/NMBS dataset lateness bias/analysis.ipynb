# NMBS Lateness bias
In this dataset, I wanted to check something peculiar. On any of the signs in the stations, the train only shows as 'late' when it has 5 or more minutes of delay. However, the NMBS holds much more accurate data of the train lateness. 

I wanted to check if there is a bias from the NMBS, where 4 minutes occurs disproportionally frequently, so as they might not need to display a train as late.

### Getting the data
The data was in the GTFS format, which proved to be somewhat tricky

```python
from google.transit import gtfs_realtime_pb2
import urllib.request
import json

feed = gtfs_realtime_pb2.FeedMessage()
response = urllib.request.urlopen('https://sncb-opendata.hafas.de/gtfs/realtime/c21ac6758dd25af84cca5b707f3cb3de')
feed.ParseFromString(response.read())
for entity in feed.entity:
    print("--------------")
    print(entity)
    print(entity.trip_update.stop_time_update[0].stop_id)    
```

I wanted to gather enough datapoints in a MongoDB database. For every train (trip), on every station (stop), I want to have the most recent delay (if any), which is denoted by a stop update. Delays of zero do not interest us for this exercise.

I immediately wanted to couple it with some other data, so as to not have to do any further data wrangling in the future. To do this, NMBS was so kind as to offer additional datasets:

```python
import pandas as pd

source_trips = pd.read_csv("gtfs_static/trips.txt")
source_agency = pd.read_csv("gtfs_static/agency.txt")
source_calendar = pd.read_csv("gtfs_static/calendar.txt")
source_calendar_dates = pd.read_csv("gtfs_static/calendar_dates.txt")
source_stop_time_overrides = pd.read_csv("gtfs_static/stop_time_overrides.txt")
source_stop_times = pd.read_csv("gtfs_static/stop_times.txt")
source_stops = pd.read_csv("gtfs_static/stops.txt")
source_routes = pd.read_csv("gtfs_static/routes.txt")
source_transfers = pd.read_csv("gtfs_static/transfers.txt")
source_translations = pd.read_csv("gtfs_static/translations.txt")
```

From this, we distill the relevant stuff. Next to that, I want to focus on the arrival time at a station, since this gives (in my opinion) a more honest view of the actual delay of a train.

```python
source_routes_u = source_routes[source_routes["route_short_name"] != "BUS"][["route_id","route_long_name"]].set_index("route_id")
source_stops_u = source_stops[["stop_id","stop_name","stop_lat","stop_lon"]].set_index("stop_id")
source_trips_u = source_trips[["trip_id","route_id"]].set_index("trip_id")
source_stop_times_u = source_stop_times[["trip_id","arrival_time","stop_id"]]
```

### Data Storage
For this exercise, I thought a MongoDB would be just fine. So I went ahead and set one up.

```python
from pymongo import MongoClient
client = MongoClient('localhost:27017')
db = client['test']
db.train_delays.delete_many({})
```

The following serves as a test to get the data format correct for importing it into MongoDB. I each time transform and import into MongoDB

```python
from pprint import pprint as pprint


feed = gtfs_realtime_pb2.FeedMessage()
response = urllib.request.urlopen('https://sncb-opendata.hafas.de/gtfs/realtime/c21ac6758dd25af84cca5b707f3cb3de')
feed.ParseFromString(response.read())
for entity in feed.entity:
        train_trip = {}
        train_trip["id"] = entity.id
        train_trip["name"] = source_routes_u.loc[source_trips_u.loc[str(entity.id)]["route_id"]]["route_long_name"]
        train_trip["delays"] = []
        for update in entity.trip_update.stop_time_update:
            if update.arrival.delay > 0:
                update_data = {}
                update_data["stop"] = source_stops_u.loc[update.stop_id]["stop_name"]
                update_data["delay"] = update.arrival.delay
                train_trip["delays"].append(update_data)
        pprint(train_trip)
        
```

Looks great, now for the logging function. This requires an extra step where it needs to check if the delay is already present. if so: overwrite it with the most current delay time. I will check this every minute.

```python
import time
db.train_delays.delete_many({})

while True:
    time.sleep(60)
    feed = gtfs_realtime_pb2.FeedMessage()
    response = urllib.request.urlopen('https://sncb-opendata.hafas.de/gtfs/realtime/c21ac6758dd25af84cca5b707f3cb3de')
    feed.ParseFromString(response.read())
    for entity in feed.entity:
        if db.train_delays.find({"id":entity.id}).count() == 0:
            print("new trip added")
            train_trip = {}
            train_trip["id"] = entity.id
            train_trip["name"] = source_routes_u.loc[source_trips_u.loc[str(entity.id)]["route_id"]]["route_long_name"]
            train_trip["delays"] = []
            db.train_delays.insert_one(train_trip)
        
        for update in entity.trip_update.stop_time_update:
            delay_time = update.arrival.delay
            stop_name = source_stops_u.loc[update.stop_id]["stop_name"]
            if db.train_delays.find({"id":entity.id, 'delays.stop':stop_name}).count() == 0 and delay_time > 0:
                print('new stop added to existing trip')
                    
                update_data = {}
                    
                update_data["stop"] = stop_name
                update_data["delay"] = delay_time
                    
                db.train_delays.update(
                    {"id":entity.id},
                    {"$push": {"delays":update_data}}
                )
                    
            elif db.train_delays.find({"id":entity.id, 'delays.stop':stop_name}).count() > 0 :
                
                current_delay = list(db.train_delays.aggregate([
                    {'$unwind':"$delays"},
                    {'$match':{"id":entity.id, 'delays.stop':stop_name}},
                    {'$project':{'delays.delay':1,'_id':0}}
                ]))[0]['delays']['delay']
                    
                    
                if delay_time != current_delay:
                    
                    print("updating a delay time")
                
                    db.train_delays.update_one(
                        {'id':entity.id,'delays.stop':stop_name},
                        {'$set':{
                            'delays.$.delay':delay_time                        
                        }}
                    )
                    
                elif current_delay == 0:
                    
                    print("removing a delay")
                    
                    db.train_delays.update( 
                        {'id':entity.id}, 
                        { '$pull': { 'delays':  {"stop":stop_name}} } )
```

### Data Analysis
First, let's get the number of delay points, and try to plot them on a graph

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

delays = [element["delays"]["delay"] for element in list(db.train_delays.aggregate([{"$unwind":"$delays"},{"$project":{"delays.delay":1}}]))]
unique, counts = np.unique(delays, return_counts=True)
print(len(counts))
plt.bar(unique[1:15],counts[1:15],width = 40)
plt.xticks(unique[1:15])
plt.xlabel("seconds delay")
plt.ylabel("number of delays")
plt.title("Number of delays per delay time")
plt.show()
```

We could try and fit a distribution on it, but just by visual reference we can observe that there seems to be no bias in favor of a 4 minute delay (240 seconds) in contrast to a 5 minute delay (300 seconds).
